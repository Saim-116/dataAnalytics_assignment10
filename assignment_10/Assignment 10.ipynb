## ***12. What Is The Distribution of Age Column Show In Graph?***


```python
# Step 1: Plot the distribution of the 'age' column using a histogram
plt.figure(figsize=(10, 6))  # Adjust the size of the plot
sns.histplot(df['age'], bins=30, kde=True, color='blue')  # kde=True adds a Kernel Density Estimate line
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
# Step 1: Using the between() method
age_between = df[df['age'].between(17, 48)]  # Finds rows where age is between 17 and 48 inclusive
total_between = age_between.shape[0]  # Get the count of rows

print(f"Total number of persons with age between 17 and 48 (inclusive) using 'between': {total_between}")

# Step 2: Without using the between() method
age_between_manual = df[(df['age'] >= 17) & (df['age'] <= 48)]  # Manual filtering with conditions
total_between_manual = age_between_manual.shape[0]  # Get the count of rows

print(f"Total number of persons with age between 17 and 48 (inclusive) without using 'between': {total_between_manual}")
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
# Step 1: Replace income values
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

# Alternatively, you can also do this using boolean indexing
# df['income'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)

# Step 2: Display the updated dataframe
print(df.head())
```

## ***18. Which Workclass getting The Highest Salary***

```python
# Step 2: Group by 'workclass' and calculate the mean income
workclass_salary = df.groupby('workclass')['income'].mean().reset_index()

# Step 3: Find the workclass with the highest average salary
highest_salary_workclass = workclass_salary.loc[workclass_salary['income'].idxmax()]

# Step 4: Display the result
print(f"Workclass with the highest average salary: {highest_salary_workclass['workclass']}")
print(f"Average Salary: {highest_salary_workclass['income']}")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
# Step 2: Calculate the total number of males and females
total_males = df[df['gender'] == 'Male'].shape[0]
total_females = df[df['gender'] == 'Female'].shape[0]

# Step 3: Calculate the number of males and females earning >50K
males_above_50k = df[(df['gender'] == 'Male') & (df['income'] == 1)].shape[0]
females_above_50k = df[(df['gender'] == 'Female') & (df['income'] == 1)].shape[0]

# Step 4: Calculate the proportion of each gender earning >50K
male_chance = males_above_50k / total_males * 100  # Percentage of males earning >50K
female_chance = females_above_50k / total_females * 100  # Percentage of females earning >50K

# Step 5: Display the results
print(f"Percentage of males earning >50K: {male_chance:.2f}%")
print(f"Percentage of females earning >50K: {female_chance:.2f}%")

# Step 6: Determine who has a better chance
if male_chance > female_chance:
    better_chance = "Males"
else:
    better_chance = "Females" if female_chance > male_chance else "Both have equal chances"

print(f"{better_chance} have a better chance of earning >50K.")
```
